```python
selected_data = complete_data.copy()

selected_data['Estimated Jackpot'] = selected_data['Estimated Jackpot'].fillna(method='ffill')
selected_data['Month'] = selected_data['Date'].dt.month

# Fix First Date
selected_data['Estimated Jackpot'] = selected_data['Estimated Jackpot'].fillna("$199 Million")

# Convert Estimated Jackpot to Numerical, multiple billion by 1000 to get millions
selected_data['Estimated Jackpot'] = selected_data['Estimated Jackpot'].apply(lambda x: x.replace('$',''))
selected_data['Estimated Jackpot'] = selected_data['Estimated Jackpot'].apply(lambda x: x.replace('Million',''))
selected_data['Estimated Jackpot'] = selected_data['Estimated Jackpot'].apply(lambda x: int(x.replace('Billion','')) * 1000 if x.find("Billion") != -1 else x)
selected_data['Estimated Jackpot'] = selected_data['Estimated Jackpot'].apply(lambda x: int(x))
```

```python
X = sampled_data[['SubCategoryName','WeekDayName','IsWeekend','Month','FranchiseID','Estimated Jackpot']]
print("X Shape:", X.shape)
# TotalSales
y = sampled_data['TotalSales']
print("Y Shape:", y.shape)
```

### Train/Test Set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=35)
print("Training Set Length:", len(X_train), "Percentage:", round((len(X_train)/len(sampled_data)),2))
print("Testing  Set Length:", len(X_test), "Percentage:", round((len(X_test)/len(sampled_data)), 2))
```

### One Hot Encoding, Convert Categorical to Numerical

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer

numeric_transformer = Pipeline(
    steps=[("imputer", SimpleImputer(strategy="median")), ("scaler", StandardScaler())]
)
categorical_transformer = OneHotEncoder(handle_unknown='ignore')
categorical_features = ['SubCategoryName','WeekDayName','IsWeekend','Month','FranchiseID']
numerical_features = ['Estimated Jackpot']

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numerical_features),
        ("cat", categorical_transformer, categorical_features),
    ]
)

prepared_data = preprocessor.fit(X)
joblib.dump(preprocessor, 'preprocessor.pkl')
```

```python
prepared_data = preprocessor.transform(X_train)
prepared_data
```

## Model Creation

```python
%%time
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor()
forest_reg.fit(prepared_data, y_train)
pickle.dump(forest_reg, open('forest_reg.pkl', 'wb'))
```

```python
X_test_prepared = preprocessor.transform(X_test)
prediction_score = forest_reg.score(X_test_prepared,y_test)
prediction_score
```
